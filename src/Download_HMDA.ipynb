### HMDA Downloads Files
This file contains code to download the HMDA and coastal data. <br/>
These files are downloaded first and separately from other downloads so that as much  memory as possible can be <br/>
available for HMDA preprocessing (HMDA_Preprocessing.ipynb) as the processing of the large HMDA files is memory intensive.<br/>

#### Required variables must be set before invoking with %run:
<strong>years</strong><br>
List of years for HMDA data to download<br>

```python
#imports
import requests
```

```python
#Setting Default parameters if no parameters specified when called to run
try:
    years
except:
    years = [2018, 2019, 2020, 2021, 2022]
```

```python
def data_directory():
    ! mkdir data
```

```python
data_directory()
```

```python
def download_Coastal():

    ! mkdir data/Coastal
    url = 'https://www2.census.gov/library/stories/2018/08/coastline-counties-list.xlsx'
    r = requests.get(url, allow_redirects=True)
    with open('data/Coastal/coastline-counties-list.xlsx', "wb") as file:
        file.write(r.content)
```

```python
download_Coastal()
```

```python
def download_HMDA(years):

    ! mkdir data/HMDA
    for year in years:
        zip_link = 'https://s3.amazonaws.com/cfpb-hmda-public/prod/snapshot-data/{}/{}_public_lar_csv.zip'.format(year,year)
        r = requests.get(zip_link, allow_redirects=True)
        with open('data/HMDA/{}.zip'.format(year), 'wb') as file:
            file.write(r.content)
```

```python
download_HMDA(years)
```
